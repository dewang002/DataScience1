# numpy select

```python
condition = [
    df['Price_euros'] > 3000,
    (df['Price_euros'] <= 3000) & (df['Price_euros'] > 2000),
    (df['Price_euros'] <= 2000) & (df['Price_euros'] > 800),
    df['Price_euros'] <=800
]
value = ['too Expensive','Expensive','Affortable','Cheap']
```

```python
df['Pricing'] = np.select(condition, value, default='unknown')
```

```python
df
```

```python
size = [
    df['Inches'] > 16,
    (df['Inches'] > 14) & (df['Inches'] <= 16),
    (df['Inches'] > 12) & (df['Inches'] <= 14),
    df['Inches'] <= 12
]
values =['very Big', 'Big', 'small', 'very small']
```
