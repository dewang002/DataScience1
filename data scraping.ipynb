```python
dic_data = {'Spanish La Liga':'SP1', 'German Bunder liga':'D1', 'English League 1':'E2'}
dic_histori = {}

root = "https://www.football-data.co.uk/mmz4281/"
for i in dic_data:
    data = [] 
    for season in range(19,21):
        df = pd.read_csv(root + str(season) + str(season + 1) + "/" + dic_data[i] +".csv")
        df.insert(1, 'season', season)
        data.append(df)
    df_concat = pd.concat(data)
    dic_histori[i] = df_concat
```
